```python
# find and remove correlated features

def correlation(dataset, threshold):
    col_corr = set()  
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]  
                col_corr.add(colname)
    return col_corr

corr_features = correlation(X_train[list(arr["Variable"])], 0.8)
print('num correlated features: ', len(set(corr_features)) )
set(corr_features)
```

```python

X_train.reset_index(inplace=True, drop=True)
X_test.reset_index(inplace=True, drop=True)
X_dev.reset_index(inplace=True, drop=True)
```

```python
import statsmodels.api as sm

def backward_elimination(data, target,significance_level = 0.01):
    
    features = data.columns.tolist()
    
    while(len(features)>0):
        
        features_with_constant = sm.add_constant(data[features])
        p_values = sm.OLS(target, features_with_constant).fit().pvalues[1:]
        max_p_value = p_values.max()
        print("max_p_value = "  ,max_p_value)
        if(max_p_value >= significance_level):
            excluded_feature = p_values.idxmax()
            print("excluded_feature = " , excluded_feature)
            features.remove(excluded_feature)
        else:
            break 
            
    return features
```
