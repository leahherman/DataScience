## Q2. How many students have each course of the English department and the total number of students in the department?

```python
#Question 2 works perfect  ......
### num students per Course
englishDepartment = 1.0
dfStudentsCoursesDept =pd.merge( classrooms[["CourseId" , "StudentId"]] , courses[["CourseId" ,"CourseName" , "DepartmentID"]]  ,on="CourseId")
dfStudentsCoursesDept.rename(columns={"DepartmentID":"DepartmentId"},inplace=True)
dfStudentsCoursesDeptA = dfStudentsCoursesDept.loc[dfStudentsCoursesDept['DepartmentId'] == englishDepartment]
dfstudentsPerEnglishCourse = dfStudentsCoursesDeptA.groupby("CourseName" ,as_index=False).agg({"StudentId":"count"})
new_names = [ "Students"  ]
old_names = ["StudentId"]
dfstudentsPerEnglishCourse.rename(columns=dict(zip(old_names, new_names)), inplace=True)

df = dfstudentsPerEnglishCourse.copy ()
totalStudents = len(dfStudentsCoursesDeptA.loc[: , ["StudentId" ]].drop_duplicates())
mycolumns = ["CourseName" , "Students"]
dfTotal = pd.DataFrame(columns=mycolumns)

lastRow = [ "total",   totalStudents ] 
dfTotal.loc[len(df)] = lastRow
dfTotal

totaldf = pd.concat([df, dfTotal])
totaldf.set_index(["CourseName" ,"Students"])
totaldf.columns
totaldf

```

## Q3. How many small (<22 students) and large (22+ students) classrooms are needed for the Science department?

```python
### num students per Course
import numpy as np

scienceDepartment = 2.0
bigClass = 23.0
dfStudentsCoursesDept =classrooms[["CourseId" , "StudentId"]].set_index('CourseId').join(courses[["CourseId"  , "DepartmentID"]] .set_index('CourseId'))
dfStudentsCoursesDept.rename(columns={"DepartmentID":"DepartmentId" ,"StudentId":"NumStudents"},inplace=True)
dfStudentsCoursesDeptA = dfStudentsCoursesDept.loc[dfStudentsCoursesDept['DepartmentId'] == scienceDepartment]
StudentPerCourse = dfStudentsCoursesDeptA.groupby("CourseId").agg({"NumStudents" : "count"})
dfResult= pd.DataFrame(columns=["Big Classes" , "Small Classes"])
dfResult.set_index(["Big Classes" , "Small Classes"])
lstSize = []
for label, content in StudentPerCourse.items():
    lst_Size = list(content)
    break 
# Array creation
array = np.array(lst_Size)
numElements = len (array)

count = 0
np.place(array, array >= bigClass, 100)
for e in array:  
    if  e ==100:
        count = count + 1


dfResult.loc[1 ,"Big Classes"] = count
dfResult.loc[1 , "Small Classes"] = numElements - count
dfResult    
```

## Q5. For which courses the percentage of male/female students is over 70%?

```python
merged_df = pd.merge(classrooms, students, on="StudentId")
df = pd.merge(merged_df, courses, on="CourseId")
type(df)
dfNumStudentsPerCoursePerGender = df.groupby(["CourseName" , "Gender" , "CourseId"] , as_index=False).agg({"StudentId" : "count"})
dfNumStudentsPerCoursePerGender.rename(columns={"StudentId":"NumStudents"},inplace=True)
dfNumStudentsPerCoursePerGenderTotal = df.groupby(["CourseName" , "CourseId"] , as_index=False).agg({"StudentId" : "count"})
dfNumStudentsPerCoursePerGenderTotal.rename(columns={"StudentId":"TotalStudents"},inplace=True)
dfNumStudentsPerCoursePerGender

for j, row in dfNumStudentsPerCoursePerGender.iterrows():
    i=dfNumStudentsPerCoursePerGenderTotal[dfNumStudentsPerCoursePerGenderTotal["CourseName"]== row["CourseName"]]
    totalVal = int(i["TotalStudents"])
    genderVal = int(row["NumStudents"])
    genderPercent = genderVal / totalVal
    #if genderPercent > o.7
    dfNumStudentsPerCoursePerGender.at[j, "TotalStudents"] = totalVal
    dfNumStudentsPerCoursePerGender.at[j, "GenderPercent"] = genderPercent
    dfNumStudentsPerCoursePerGender.at[j, "AbovePercent"] =  (" ", "**")[genderPercent > 0.7 ]   

dfNumStudentsPerCoursePerGender
dfResult=dfNumStudentsPerCoursePerGender[dfNumStudentsPerCoursePerGender["AbovePercent"]== "**"]
dfResult
```

## Q6. For each department, how many students passed with a grades over 80?

```python
shmonim = 80.0
zero = 0.0
merged_df = pd.merge(classrooms, courses, on="CourseId")
merged_df.rename(columns={"DepartmentID":"DepartmentId"},inplace=True)
df = pd.merge(merged_df, departments, on="DepartmentId")
 
dfAllStudents = df[["DepartmentName" , "StudentId"]].drop_duplicates().groupby(["DepartmentName"] ,as_index=False).agg({"StudentId" :"count"})
dfAllStudents.rename(columns={"StudentId":"allStudents"},inplace=True)


#CASE WHEN (a.degree > 80) THEN (1) ELSE (0) END AS gt80
for j, row in df.iterrows():
    degree = int(row["degree"])
    df.at[j, "gt80"] = (0.0,1.0)[degree > shmonim]

dfRelavant = df[df["gt80"]  >  zero]
dfRelavant = df[df["degree"]  >  shmonim]
dfResult = dfRelavant[["DepartmentName" , "StudentId"  ]].drop_duplicates().groupby(["DepartmentName"] ,as_index=False).agg({"StudentId" :"count"})


dfResult.rename(columns={"StudentId":"students80"},inplace=True)
dfResult 
for j, row in dfResult.iterrows():
    name = row["DepartmentName"]
    i=dfAllStudents[dfAllStudents["DepartmentName"]== name ]
    totalVal = int(i["allStudents"])
    dfResult.at[j, "allStudents"] = totalVal
    dfResult.at[j, "percentStudents"] =row["students80"]/ totalVal

dfResult
```

## Q7. For each department, how many students passed with a grades under 60?

```python
shishim = 60.0
zero = 0.0
merged_df = pd.merge(classrooms, courses, on="CourseId")
merged_df.rename(columns={"DepartmentID":"DepartmentId"},inplace=True)
df = pd.merge(merged_df, departments, on="DepartmentId")
 
dfAllStudents = df[["DepartmentName" , "StudentId"]].drop_duplicates().groupby(["DepartmentName"] ,as_index=False).agg({"StudentId" :"count"})
dfAllStudents.rename(columns={"StudentId":"allStudents"},inplace=True)
#  to check later  df.groupby(['DepartmentName' ]).StudentId.nunique()


dfRelavant = df[df["degree"]  <  shishim]
dfResult = dfRelavant[["DepartmentName" , "StudentId"  ]].drop_duplicates().groupby(["DepartmentName"], as_index=False).agg({"StudentId" :"count"})


dfResult.rename(columns={"StudentId":"students60"},inplace=True)
dfResult 
for j, row in dfResult.iterrows():
    name = row["DepartmentName"]
    i=dfAllStudents[dfAllStudents["DepartmentName"]== name ]
    totalVal = int(i["allStudents"])
    dfResult.at[j, "allStudents"] = totalVal
    dfResult.at[j, "percentStudents"] =row["students60"]/ totalVal

dfResult

```

## Q8. Rate the teachers by their average student's grades (in descending order).

```python

merged_df = pd.merge(classrooms, courses, on="CourseId")
df = pd.merge(merged_df, teachers, on="TeacherId")
 
dfTeachers = df[["TeacherId" , "FirstName" , "LastName" , "degree" ]].drop_duplicates().groupby(["FirstName" ,"LastName" ] ,as_index=False).agg({"degree" :"mean"})
dfResult =dfTeachers.sort_values(by="degree" ,  ascending=False )
dfResult.rename(columns={"degree":"avgDegree"},inplace=True)
dfResult["Teacher"] = dfResult["FirstName"] + "  " + dfResult["LastName"]
print(dfResult[["Teacher" , "avgDegree"]])
```
